# Gaming and Mental Health Dataset Analysis

Exploratory Data Analysis (EDA) of a large dataset about gaming behavior and mental health.

Dataset source:
https://www.kaggle.com/datasets/sharmajicoder/gaming-and-mental-health

## 1. Import libraries

Import the neccessary Python libraries for data loading and analysis.

```python
import kagglehub
import os
import pandas as pd
```

## 2. Descargar dataset

Download the latest version of the dataset directly from Kaggle.

```python
path = kagglehub.dataset_download("sharmajicoder/gaming-and-mental-health")

print("Path to dataset files:", path)
```

## 3. Inspect Dataset Files

List the files included in the dataset directory to identify the CSV file.

```python
os.listdir("/kaggle/input/gaming-and-mental-health")
```

## 4. Load Dataset

Load the dataset into a Pandas Dataframe for further analysis.

```python
df = pd.read_csv("/kaggle/input/gaming-and-mental-health/gaming_mental_health_10M_40features.csv")
df.head()  # muestra las primeras filas
```

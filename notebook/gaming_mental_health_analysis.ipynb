## 6. Data Distribution Analysis
Understanding how key variables are distributed

### 6.1 Create a Smaller Sample for Faster Exploration

Create a smaller subset of the dataset for quicker experimentation and analysis.

```python
df_sample = df.sample(10000, random_state=42)
df_sample
```

### 6.2 Age Distribution
Analyze the distribution of players' ages

```python
plt.figure(figsize=(8,5))
sns.histplot(df_sample["age"], bins=30, kde=True)

plt.title("Age Distribution of Players")
plt.xlabel("Age")
plt.ylabel("Count")

plt.show()
```

Findings
- The dataset includes players ranging approximately from 13 to 59 years old.
- The distribution appears fairly uniform, suggesting that the dataset was likely generated or sampled to represent players across many age groups.
- There is no strong concentration in a specific age range.

### 6.3 Gaming Hours Distribution
Analyze the distribution of players' daily gaming hours

```python
plt.figure(figsize=(8,5))
sns.histplot(df_sample["daily_gaming_hours"], bins=30, kde=True)

plt.title("Gaming Hours Distribution")
plt.xlabel("Daily Gaming Hours")
plt.ylabel("Number of Players")

plt.show()
```

Findings
- Most players spend between 1 and 5 hours gaming per day.
- The distribution is right-skewed, meaning a small number of players spend significantly more time gaming.
- A few extreme cases show players gaming more than 15 hours per day, which may represent heavy gamers.

### 6.4 Correlation Analysis
Analyze the relationships between gaming habits, mental health scores, and lifestyle factors.

```python
corr_cols = [
    "daily_gaming_hours",
    "sleep_hours",
    "stress_level",
    "anxiety_score",
    "depression_score",
    "exercise_hours",
    "screen_time_total",
    "addiction_level"
]

plt.figure(figsize=(8,6))

sns.heatmap(
    df_sample[corr_cols].corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")

plt.show()
```

Findings
- Strong Positive Correlations: daily_gaming_hours shows strong correlations with addiction_level (0.89) and screen_time_total (0.72).
- Weak Mental Health Relationships: Variables such as stress_level, anxiety_score, and depression_score show near-zero correlation with gaming hours.
- Lifestyle Independence: exercise_hours and sleep_hours show little to no correlation with gaming or mental health metrics.

### 6.5 Gaming Hours vs Addiction Level

Analyze the relationship between gaming time and addiction level.

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x=df_sample["daily_gaming_hours"],
    y=df_sample["addiction_level"],
    alpha=0.4
)

plt.title("Gaming Hours vs Addiction Level")
plt.xlabel("Daily Gaming Hours")
plt.ylabel("Addiction Level")

plt.show()
```

Findings
- Strong Positive Relationship: There is a clear positive relationship between daily gaming hours and addiction level, confirming the high correlation observed in the correlation analysis.
- Addiction Ceiling Effect: Players who exceed roughly 12–15 hours of gaming per day often reach the maximum addiction score (10).
- Lower Addiction at Moderate Play: Most players who game between 0 and 5 hours per day show relatively low addiction scores.

## 7. Key Insights
Summary of the most important findings discovered during the exploratory data analysis.

- Moderate Gaming Behavior: Most players spend between 1 and 5 hours per day gaming, indicating that extreme gaming habits are relatively uncommon in the dataset.
- Gaming Time and Addiction: A strong positive correlation exists between daily gaming hours and addiction level, suggesting that increased gaming time is closely associated with higher addiction scores.
- Weak Links to Mental Health Metrics: Variables such as stress, anxiety, and depression show little to no correlation with gaming hours in this dataset.
- Lifestyle Factors Appear Independent: Sleep and exercise habits do not show meaningful linear relationships with gaming behavior or mental health scores.
